```python
x=[1,2,3,4,5]
y=[3,54,6,98,50]
# plt.plot(x,y,marker ="o")   # it is use to circle the edge
plt.plot(x,y,marker ="o",ms=10,             # ms= it is use to size of circlr
            mfc="red",mec="green",         # mfc--> ti change face color of edge  mec--> use for edge   edge color
                color="black",
              lw=3,ls= ":",
              alpha=0.7,label = "limeplot")         # alpha ==  0--1     0--> hilghlite        1--> dark
plt.title("limeplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
# scaterplot        ##     in this only edge point show           all functionality same as line plot
plt.scatter(x,y)

plt.show()
```

```python
# histogram --> distribution of data in interval(bins)
arr = np.random.randint(20000,40000,100)
plt.hist(arr,bins=25)
plt.show()
```

```python
## pie char --> percentage
x = ["Madhya Pradesh","Uttar Pradesh","Rajasthan","Gujrat"]
y=[10,20,30,40]
explodes=(0.1,0,0,0)
plt.pie(y,labels=x,autopct="%1.2f%%",shadow=True,explode=explodes)        # autopct is used to show percentage
plt.legend()   
plt.savefig("piechart.jpeg")                                                                     # explodes is use to seprate one part
plt.show()
```

```python
  # subplot
x = np.linspace(0,10,100)
plt.Figure(figsize=(10,6))
plt.subplot(2,2,1)     # 2 rows, 2 columns ,1st graph
plt.plot(x,np.sin(x))
plt.title("sin")

plt.subplot(2,2,2)     # 2 rows, 2 columns ,1st graph
plt.plot(x,np.cos(x))
plt.title("cos")

plt.subplot(2,2,3)     # 2 rows, 2 columns ,1st graph
plt.plot(x,np.tan(x))
plt.title("tan")

plt.subplot(2,2,4)     # 2 rows, 2 columns ,1st graph
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()        # it is used for seprate overlape condition
plt.show()
import warnings
warnings.filterwarnings("ignore")
```
