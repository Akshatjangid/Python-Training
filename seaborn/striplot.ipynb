strip plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.stripplot(var)
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",edgecolor="b",size=5,alpha=0.3,marker="<")        #  jitter=0
plt.show()
```
