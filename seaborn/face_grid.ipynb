```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
fg=sns.FacetGrid(var,col="sex",hue="day")                    # making two graph we use col==""
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex")                    # making two graph we use col==""
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day",hue="sex",palette="summer")                    # making two graph we use col==""
fg.map(plt.bar,"total_bill","tip",edgecolor="b").add_legend()
plt.show()
```
