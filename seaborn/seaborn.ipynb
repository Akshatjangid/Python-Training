```python
#seaborn--> is a data visualization tool for statistical graphics
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
var= [1, 3, 2, 5, 4]
var_1= [2, 4, 6, 8, 6]

plt.plot(var,var_1)
plt.show()
```

```python
sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
y_1=sns.load_dataset("penguins").head(50)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",
             palette="Accent",markers=["o",">"],dashes=False,legend=False)    #Hue --> is use to make a graph respect to popular data like sex etc....
# palette is use to change the color of graph       # marker --> is use to edge poin style                   # size--> is use to change graph size
# dashes is usse to true or false in false the female doted line are single line
plt.grid()
plt.title("Seaborn")
plt.show()
```
