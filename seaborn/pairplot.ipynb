pair plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips").head(20)
var
```

```python
# sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="Set3")
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],palette="Set3",x_vars=["total_bill","tip"],kind="kde")            #kind--> reg , scatter , kde , hist
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],markers=["<","*"])            #kind--> reg , scatter , kde , hist
plt.show()
```
