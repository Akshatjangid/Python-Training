voilin plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,palette="RdPu",linecolor="b")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Lunch","Dinner"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,inner="quart")
plt.show()
```

```python
sns.violinplot(y="day",x="total_bill",data=var,hue="sex",split=True)
plt.show()
```
