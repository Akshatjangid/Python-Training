```python
m={"Male":"o","Female":">"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(100,40),palette="Dark2",alpha=1,markers=m)
```
