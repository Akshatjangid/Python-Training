Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",order=["Fri","Sat","Thur","Sun"])
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y="day",x="total_bill",data=var,hue="sex",order=["Fri","Sat","Thur","Sun"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"y"},linewidth=2,palette="Accent",orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y="day",x="total_bill",data=var,hue="sex",order=["Fri","Sat","Thur","Sun"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"y"},
            linewidth=2,palette="Accent")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"y"},linewidth=2,palette="Accent",orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=var["tip"])
```
