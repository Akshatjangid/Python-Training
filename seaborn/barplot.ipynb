```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var=sns.load_dataset("penguins")
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.set(style="dark")
# sns.barplot(x=var.island,y=var.bill_depth_mm)   errwidth=6
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",order=order_1,hue_order=["Female","Male"],errorbar=('ci', 100),saturation=0.1,err_kws={'color': 'g'},capsize=0.7,
            dodge=True)       # if nit use var.1 we use string and use data 
# ci--> is use to change interval of the data                                       # hue--> is use to compare an item with respect to something

plt.show()
```
